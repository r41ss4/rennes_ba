# Diagnostic Analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore, ttest_ind, chi2_contingency
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/r41ss4/rennes_ba/refs/heads/main/data/merged/merged_data.csv')
df
```

### Identify Anomalies

```python
# Detect outliers and unusual patterns
def detect_outliers(df, column):
    z_scores = zscore(df[column])
    return df[np.abs(z_scores) > 3]
```

```python
# 篩選數值型欄位
numeric_columns = df.select_dtypes(include=[np.number]).columns

# 儲存每個欄位的異常值
outliers_dict = {}

# 迭代檢測每個數值型欄位
for column in numeric_columns:
    # 計算 Z 分數
    z_scores = zscore(df[column])
    outliers = df[np.abs(z_scores) > 3]  # 異常值判定條件
    
    if not outliers.empty:
        outliers_dict[column] = outliers
        print(f"Outliers detected in column '{column}':")
        print(outliers)
        
        # 繪製箱型圖，突出顯示異常值
        plt.figure(figsize=(8, 6))
        sns.boxplot(x=df[column])
        plt.title(f"Box Plot of {column} (Outliers Highlighted)")
        plt.show()
    else:
        print(f"No outliers detected in column '{column}'.")

# 總結檢測結果
if outliers_dict:
    print("\nSummary of all outliers detected:")
    for column, outlier_rows in outliers_dict.items():
        print(f"Column: {column}, Outlier Count: {len(outlier_rows)}")
else:
    print("\nNo outliers detected in any numeric column.")
```

```python
# Calculate thresholds for outliers using IQR (Interquartile Range)
Q1 = df['rating'].quantile(0.25)
Q3 = df['rating'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['rating'] < lower_bound) | (df['rating'] > upper_bound)]
print(f"Number of Outliers Detected: {len(outliers)}")
print(outliers[['title', 'rating', 'certificate']].head())

# Plotting to visualize
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='rating')
plt.title("Boxplot of Ratings (Outliers Detection)")
plt.xlabel("Rating")
plt.show()
```

```python
from scipy.stats import shapiro

# Perform Shapiro-Wilk test for normality on ratings
stat, p_value = shapiro(df['rating'])

print(f"Shapiro-Wilk Test: stat = {stat:.3f}, p-value = {p_value:.3f}")
if p_value < 0.05:
    print("The ratings are not normally distributed (significant deviation).")
else:
    print("The ratings are normally distributed.")
```

### Correlation Analysis

```python
# 指定需要分析的欄位
selected_columns = ['duration_min', 'rating', 'votes', 'seasons']

# 過濾選定欄位中的非數值型資料
numeric_columns = df[selected_columns].select_dtypes(include=['number']).columns
numeric_df = df[numeric_columns]

# 計算相關矩陣
correlation_matrix = numeric_df.corr()
print(f"Correlation Matrix:\n{correlation_matrix}")
```

```python
# Create heatmaps to visualize correlations
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### Root Cause Analysis

```python
# Create a pivot table to analyze average rating and votes by genre
pivot_table = df.pivot_table(
    values=['rating', 'votes'], 
    index='certificate', 
    aggfunc={'rating': 'mean', 'votes': 'mean'}
)

# Remove 'Unknown' and 'Unrated'
filtered_pivot_table = pivot_table.drop(index=['Unknown', 'Unrated'], errors='ignore')
```

```python
# Display the filtered pivot table
print("Filtered Pivot Table (Certificate vs Rating and Votes):")
print(filtered_pivot_table)
```

### Comparative Analysis

```python
# Boxplot to compare ratings across certificates

# Remove rows where 'certificate' is 'Unknown' or 'Unrated'
filtered_data = df[~df['certificate'].isin(['Unknown', 'Unrated'])]

plt.figure(figsize=(10, 6))
sns.boxplot(data=filtered_data, x='certificate', y='rating')
plt.title('Ratings by Certificate')
plt.xlabel('Certificate')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.show()                                                                                                                   
```

### Hypothesis Testing

```python
from scipy.stats import ttest_ind

# Filter data for two specific certificates
group1 = df[df['certificate'] == 'TV-MA']['rating']
group2 = df[df['certificate'] == 'TV-14']['rating']

# Perform t-test
t_stat, p_value = ttest_ind(group1, group2, nan_policy='omit')
print(f"T-Test Results: t-stat = {t_stat:.3f}, p-value = {p_value:.3f}")

if p_value < 0.05:
    print("The difference in ratings between the two certificates is statistically significant.")
else:
    print("No significant difference in ratings between the two certificates.")
```
