# Clean the dataset

```python
clean_dataframe = dataframe
```

```python
clean_dataframe.drop_duplicates()
```

```python
clean_dataframe.isna().sum()
```

```python
# For 'description' we'll change it to 'No description available' to avoid future mistakes
clean_dataframe['description'] = dataframe['description'].fillna('No description available')
```

```python
clean_dataframe.isna().sum()
```

```python
#For numeric columns, the median will be used so that it doesn't change the important measures 
num_cols = ['imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score']
for col in num_cols:
    clean_dataframe[col] = clean_dataframe[col].fillna(clean_dataframe[col].median())
```

```python
clean_dataframe.isna().sum()
```

```python
# 'Seasons' will be cleaned by assuming a 0 for movies or no more seasons
clean_dataframe['seasons'] = clean_dataframe['seasons'].fillna(0).astype(int)
```

```python
clean_dataframe.isna().sum()
```

```python
# For cleaning 'age_certification' the blank rows will be filled with 'Unknown'
clean_dataframe['age_certification'] = clean_dataframe['age_certification'].fillna('Unknown')
```

```python
clean_dataframe.isna().sum()
```

```python
# It is important to convert the lists saved as strings as real lists to facilitate further analysis
columns = ['genres', 'production_countries']
for col in columns:
    clean_dataframe[col] = dataframe[col].apply(lambda x: eval(x) if isinstance(x, str) else x)
```

# Normalize Data 

```python

# In order to normalize the data, we'll limite the release year to reasonable years (from 1900-2024)
clean_dataframe = clean_dataframe[(clean_dataframe['release_year'] >= 1900) & (clean_dataframe['release_year'] <= 2024)]
```

```python
clean_dataframe
```

```python
# We're also limiting the values that are out of the range in 'runtime'
clean_dataframe['runtime'] = dataframe['runtime'].apply(lambda x: x if x > 0 and x < 500 else None)
clean_dataframe['runtime'] = dataframe['runtime'].fillna(dataframe['runtime'].median())
    
```

```python
clean_dataframe
```

```python
#And just to be sure, we're limiting the values out of range in scores
clean_dataframe['imdb_score'] = dataframe['imdb_score'].apply(lambda x: x if 0 <= x <= 10 else None)
clean_dataframe['tmdb_score'] = dataframe['tmdb_score'].apply(lambda x: x if 0 <= x <= 10 else None)
```

```python
clean_dataframe
```

```python
 # It is important to also standarize the strip & lowercase formats
clean_dataframe['title'] = dataframe['title'].str.strip()
clean_dataframe['type'] = dataframe['type'].str.strip().str.upper()
clean_dataframe['age_certification'] = dataframe['age_certification'].str.strip()
```

```python
clean_dataframe
```

 # Add columns that could be useful


```python
#In this case a new column will be created to calculate the age of the movie based on the released year and the actual year. 
#This could be useful for a next analysis
from datetime import datetime
current_year = datetime.now().year
clean_dataframe['title_age'] = current_year - dataframe['release_year']
```
